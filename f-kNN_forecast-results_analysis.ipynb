```python
import pandas as pd
import numpy as np

path_to_data = '/Users/Guille/Desktop/dynamic_update/data/'
```

```python
file_name = 'wind-Aguayo_Wind-124.csv'

rs_ = pd.read_csv(path_to_data + file_name)

scores_ = ['IS975', 'IS95', 'IS90', 'IS80', 'IS60', 
           'CS975', 'CS95', 'CS90', 'CS80', 'CS60', 
           'PITmean', 'PITstd', 'LogS', 'CRPS', 'WIS']

meta_ = ['index', 'forget_rate', 'update_rate', 'length_scale', 
         'edf_prob', 'lamdba', 'smoothing', 'resource', 'asset', 
         'year_day', 'interval']

df_ = []
for i in rs_['index'].unique():
    
    rs_p_ = rs_.loc[rs_['index'] == i]
    if not rs_p_.isnull().values.any():
        df_.append(pd.concat([rs_p_[meta_].iloc[0], rs_p_[scores_].agg('mean').T], axis = 0))

df_ = pd.concat(df_, axis = 1).T

print(df_.sort_values(['LogS']))

forget_rate_  = [1, 5, 10]
update_rate_  = [0, 1000, 10000]
lamdba_       = [0, 10, 1000]
edf_prob_     = [0.001, 0.0005, 0.0001]
length_scale_ = [10, 100, 1000]
```
